```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
season_no_lst = []
episode_no_lst = []
episode_name_lst = []
episode_des_lst = []
episode_date_lst = []
episode_score_lst = []
episode_votes_lst = []

for season in range(6):
    
    season_no = season + 1

    print(f'>> parsing season: {season_no}')

    url = f'https://www.imdb.com/title/tt4052886/episodes?season={season_no}'
    response = requests.get(url)
    season_page = BeautifulSoup(response.content)

    episode_tiles=season_page.findAll('div',attrs={'class':'info'})

    for episode_no,episode in enumerate(episode_tiles):

        episode_name=episode.strong.a.text
        print(f'episode_name: {episode_name}')

        episode_des = episode.find('div',attrs={'class':'item_description'}).text.strip('\n')
        print(f'episode_des: {episode_des}')

        episode_date = episode.find('div',attrs={'class':'airdate'}).text.strip().replace('.'," ")
        print(f'episode_date: {episode_date}')

        episode_score= episode.find('span',attrs={'class':'ipl-rating-star__rating'}).text
        print(f'episode_score : {episode_score}')

        episode_votes = episode.find('span',attrs={'class':'ipl-rating-star__total-votes'}).text.strip('()')
        print(f'episode_votes : {episode_votes}')

    
        print('\n')

        season_no_lst.append(season_no)
        episode_no_lst.append(episode_no + 1)
        episode_name_lst.append(episode_name)
        episode_des_lst.append(episode_des)
        episode_date_lst.append(episode_date)
        episode_score_lst.append(episode_score)
        episode_votes_lst.append(episode_votes)


print('\n')
```

```python
data= pd.DataFrame({'season_no':season_no_lst,
                    'episode_no':episode_no_lst,
                    'episode_name':episode_name_lst,
                    'episode_des':episode_des_lst,
                    'episode_airdate':episode_date_lst,
                    'episode_rating':episode_score_lst,
                    'episode_votes':episode_votes_lst})
```
